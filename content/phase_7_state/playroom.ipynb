## 🚦 7.1: Directed Transition Graphs
**Goal:** Build a pure Python router state machine.

In programmatic governed inference, code governs which transitions are active. Let's build a static routing map using a dictionary graph.

```python
routing_graph = {
    'START': ['GET_INPUT'],
    'GET_INPUT': ['VALIDATE', 'ERROR'],
    'VALIDATE': ['EXECUTE', 'GET_INPUT'],
    'EXECUTE': ['END'],
    'ERROR': ['GET_INPUT']
}

def is_transition_legal(current: str, target: str) -> bool:
    return target in routing_graph.get(current, [])

# Verify transitions:
```
