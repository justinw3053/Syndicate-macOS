## 💻 P.0: Environment Navigation & Shell Mechanics
### Conceptual Lecture: The Sandbox Boundaries
A computer runs code by executing a binary inside a specific **environment path**. If you run scripts blindly without understanding where your compiler is looking for packages, you will encounter the dreaded `ModuleNotFoundError`.

A **Virtual Environment (`venv`)** isolates your packages by copying the Python interpreter into a local folder and modifying the environment's `PATH` variable so imports point to local libraries rather than global ones. 

Let's programmatically prove that we are running inside our active virtual environment sandbox.

```python
import sys

# In Python, standard path isolation can be verified by checking if sys.prefix (active runtime) 
# differs from sys.base_prefix (global system python install).
is_venv = sys.prefix != sys.base_prefix

print(f"Active Python Path: {sys.executable}")
print(f"Running inside Virtual Environment: {is_venv}")

assert is_venv == True, "CRITICAL: You are running on the system python! Sourced venv first!"
```
