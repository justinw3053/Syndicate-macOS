## ⚙️ P.4: Functions & Scope
### Conceptual Lecture: Scope Frames and Memory Isolation
Every time a Python function is executed, it creates a new **local frame (namespace)** on the Call Stack. Any variable assigned inside this function exists *only* within this local frame. When the function returns or finishes executing, its local frame is popped off the stack and destroyed, and all its internal local variables vanish.

To return values computed inside the function back to the calling scope, you must use the `return` keyword. Without an explicit `return`, a function automatically returns `None`.

#### Mutable Argument Side-Effects
When you pass a **mutable object** (like a `list` or `dict`) to a function, Python passes a reference to the *same* object in memory. If you modify that object inside the function (e.g. `lst.append(5)`), the changes will be visible outside the function! This is called a **side-effect**. To write pure, mathematical functions, we often avoid side-effects by creating a new copy of the data inside the function instead of mutating the input.
