## 📦 P.1: Variables & Accumulators
### Conceptual Lecture: Name-Binding and the Physics of Memory
In Python, variables are not "boxes" that hold values. They are **symbolic names (references)** bound to objects in memory.
When you write `x = 5`, Python allocates memory for the integer object `5` and binds the name `x` to it.
If you write `x = x + 1`, Python evaluates `x + 1` (creating a new integer object `6`) and re-binds the name `x` to this new object. The original object `5` is discarded.

### 🧮 Math-to-Python Translation Key: Summation & Subscripts
Before we look at the code, let's decode the math notations. They look scary, but they are just standard Python loops:

1. **The Greek Sigma ($\sum_{i=1}^{N}$)**: This is just a **`for` loop**!
   - The symbol $\sum$ (Sigma) stands for **"Sum"**.
   - The bottom $i=1$ is where the loop starts (`for i in range(1, ...)`).
   - The top $N$ is where the loop stops.
   - The expression to the right of the Sigma (e.g., $i$) is what you add to your accumulator on each step.
   - *Python Equivalent:*
     ```python
     total = 0
     for i in range(1, N + 1):
         total += i
     ```

2. **Subscripts ($S_i$ and $S_{i-1}$)**: These are just **List Indices**!
   - $S_i$ is the state at index `i` (like `S[i]`).
   - $S_{i-1}$ is the state at the *previous* index `i-1` (like `S[i-1]`).
   - So, $S_i = S_{i-1} + i$ just means: *"The new total at step `i` is the old total at step `i-1` plus `i`."*
   - In Python, we optimize memory by overwriting a single variable instead of storing a history list: `total = total + i`.

### 🚀 Active Lab Exercise: The Factorial Accumulator
The factorial of $N$ ($N!$) is the product of all positive integers less than or equal to $N$:

$$\text{Product Pi: } \quad N! = \prod_{i=1}^{N} i = 1 \times 2 \times 3 \times \dots \times N$$

Which can be written as a recurrence relation:

$$F_i = F_{i-1} \times i \quad \text{where } F_0 = 1$$

**🧮 Math-to-Python Translation Key:**
- The Pi symbol ($\prod$) works exactly like the Sigma ($\sum$) sum symbol, except it performs **multiplication** (`*`) instead of addition (`+`) on each step!
