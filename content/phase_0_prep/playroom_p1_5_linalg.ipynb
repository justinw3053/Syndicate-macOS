## P.1.5: Linear Algebra & Matrix Primitives
### Conceptual Lecture: Representing Multidimensional Space
In machine learning, data exists in high-dimensional coordinates. We don't use heavy frameworks like NumPy first; we use **pure Python standard primitives** to master the arithmetic "physics."

- A **Vector** is represented as a single list of floats: `[1.0, 2.0, 3.0]`.
- A **Matrix** is represented as a list of lists (a nested coordinate grid): `[[1.0, 0.0], [2.0, 1.0]]` where each sub-list is a row.

### Math-to-Python Translation Key: Dot Product & Multiplications
1. **Dot Product ($\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{N} a_i b_i$)**:
   - Multiplication each corresponding coordinate index together ($a_1 \times b_1$, $a_2 \times b_2$, etc.) and sum the results.
   - *Mechanic (Long-form):* Loop through indices, multiply pairs, and add to an accumulator.
   - *Python Sugar:* `sum(x * y for x, y in zip(a, b))`

2. **Matrix Multiplication ($\mathbf{C}_{ij} = \sum_{k=1}^{K} \mathbf{A}_{ik} \mathbf{B}_{kj}$)**:
   - The cell at row $i$ and column $j$ in the output matrix $\mathbf{C}$ is computed by taking the **dot product** of Row $i$ of Matrix $\mathbf{A}$ against Column $j$ of Matrix $\mathbf{B}$.
   - To easily multiply row-by-column, we can **Transpose** Matrix $\mathbf{A}$ so its columns become rows!
   - *Transpose Mechanic:* Loop through columns of B, and build a new row from each column.
   - *Transpose Sugar:* `cols_b = [[row[i] for row in B] for i in range(len(B[0]))]`

#### The Mathematical Necessity
A **Dot Product** measures the overlap of two coordinates: 
$$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{N} a_i b_i$$

A **Matrix Multiplication** multiplying Matrix $\mathbf{A}$ ($M \times K$) and Matrix $\mathbf{B}$ ($K \times N)) computes the dot product of each row of $\mathbf{A}$ against each column of $\mathbf{B}$:
$$\mathbf{C}_{ij} = \sum_{k=1}^{K} \mathbf{A}_{ik} \mathbf{A}_{kj}$$

#### The Rosetta Stone
We will first calculate this using explicit `for` loops to see the mechanics. Only after proving the loop works will we compress it into Pythonic one-liners using `zip()`M or list comprehensions.
