## 🔀 P.2: Conditionals & Control Flow
### Conceptual Lecture: Branching Gates and Order of Evaluation
Conditionals act as logical gates. When Python encounters an `if` statement, it evaluates the boolean expression. If it is `True`, it enters that block and **completely ignores** all subsequent `elif` or `else` branches tied to that gate. 

If we use multiple independent `if` statements instead of `elif` gates, Python evaluates **every single one of them** sequentially, regardless of whether prior gates were open. This causes overlapping evaluation bugs.

### 🧪 Discovery Lab: The Overlapping Gates
Suppose we want to classify a score into grading bands: "Gold" ($\ge 90$), "Silver" ($\ge 70$), and "Bronze" ($\ge 50$).

````python
score = 95
if score >= 50:
    print("Bronze")
if score >= 70:
    print("Silver")
if score >= 90:
    print("Gold")
# Watch the output. Why does it print multiple grades for a single score?
````

#### The Logical Necessity
We need **mutual exclusivity**. Selecting one branch must automatically close off and skip all alternative paths. In Python, this is achieved by chaining conditions using `elif` and `else` gates, structured in order of highest specificity first.

#### The Rosetta Stone
```python
if score >= 90:
    grade = "Gold"
elif score >= 70:
    grade = "Silver"
elif score >= 50:
    grade = "Bronze"
else:

```
